```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

## 데이터 관계 시각화
* 관계도: 여러 데이터 사이의 관계를 살펴보기 위한 그래프

### 히트맵(heatmap)
* 데이터 간의 관계를 **색상**으로 표현한 그래프
* 비교해야할 데이터가 많을때 주로 사용
* `sns.heatmap(...)`; 함수를 사용해서 호출

```python
# 비행기 탑승자 수 데이터
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.head
```

```python
flights.tail
```

```python
flights.info()
```

```python
flights.describe()
```

```python
# 각 연도의 월별 승객수
flights.pivot(index='month', columns='year', values='passengers')
fp = flights.pivot(index='month', columns='year', values='passengers')
fp
```

```python
sns.heatmap(data=fp) 
```

```python
sns.heatmap(data=fp, cmap='mako')
```

```python
sns.heatmap(data=fp, cmap='coolwarm') # 많이 사용
```

### 라인플롯(lineplot)
* 두 수치형 데이터 사이의 관계를 나타낼 때 사용(직선)
* x에 전달한 값에 따라 y에 전달한 평균 + 신뢰구간
* `sns.lineplot()`

```python
sns.lineplot(x='year', y='passengers', data=flights)
```

### 산점도(scatterplot)
* 두 데이터간의 관계를 점으로 표현한 그래프

```python
tips = sns.load_dataset("tips")
```

```python
tips.head
```

```python
tips.tail
```

```python
tips.info()
```

```python
tips.describe()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day')
```

### 회귀선을 포함한 산정도 그래프(regplot)
* `regplot()`: 산정도 + 선형회귀선
* 회귀선을 그리면 전반적인 상관관계 파악이 쉬움

```python
sns.regplot(x='total_bill',y='tip', data=tips) # 95% 신뢰구간
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, ci=99) # 99% 신뢰구간
```

```python
sns.pairplot(tips)  # 모든 방향의 그래프로 그림
```
